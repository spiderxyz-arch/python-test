```python
import numpy as np 
def sigmoid(x):
    return 1/(1+np.exp(-x))
def predict_spam_or_notspam(words):
    spam = words -300
    probability = sigmoid(spam)
    if probability >=0.5:
       status = 'not spam'
    else:
        status = 'spam' 
    return probability,status
email = [100,250,300]
for words in email:
    result = predict_spam_or_notspam(words)
    proability = result[0]
    status = result[1]
    print(proability)
    print(status)
```
